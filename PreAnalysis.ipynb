```python
activated_jobs = pd.read_csv('Rapporti_di_lavoro_attivati.csv')
#lavori_attivati.head()
```

# Cleaning and Transformation Hiring DataSet

```python
activated_jobs_cleaning = activated_jobs.copy()
activated_jobs_cleaning
```

## Drop Duplicates and Null Values

```python
# Drop duplicates values
activated_jobs_cleaning = activated_jobs.drop_duplicates()
```

```python
activated_jobs_cleaning.isnull().sum()
```

```python
# Drop Null values
activated_jobs_cleaning = activated_jobs.dropna()
```

```python
# Creation Column TIPO
activated_jobs_cleaning["TIPO"]="ATTIVATO"
```

```python
# Only two Provinces from Dataset are selected
activated_jobs_cleaning = activated_jobs_cleaning.loc[(activated_jobs_cleaning['PROVINCIAIMPRESA'] == 'MILANO') |
                                                      (activated_jobs_cleaning['PROVINCIAIMPRESA'] == 'BERGAMO')]
activated_jobs_cleaning.head()
```

## Fixing Year Range

```python
activated_jobs_cleaning.head()
```

```python
# Transform column DATA from object to datetime
activated_jobs_cleaning['DATA'] = pd.to_datetime(activated_jobs_cleaning['DATA'])
```

```python
activated_jobs_cleaning.dtypes
```

```python
# Creation column ANNO to select specific interval times 
activated_jobs_cleaning['ANNO'] = activated_jobs_cleaning.DATA.dt.year
activated_jobs_cleaning.head()
```

```python
# Time interval expressed in ANNO  
interval_year = (activated_jobs_cleaning['ANNO'] > 2007) & (activated_jobs_cleaning['ANNO'] <= 2021)
activated_jobs_cleaning = activated_jobs_cleaning.loc[interval_year]
```

```python
# Selection of Computer Science sectors
activated_jobs_cleaning = activated_jobs_cleaning[activated_jobs_cleaning["SETTOREECONOMICODETTAGLIO"].str.contains("informatica") == True]
activated_jobs_cleaning["SETTOREECONOMICO"] = activated_jobs_cleaning["SETTOREECONOMICODETTAGLIO"]
activated_jobs_cleaning = activated_jobs_cleaning.drop(["SETTOREECONOMICODETTAGLIO"], axis = 1)
activated_jobs_cleaning["SETTOREECONOMICO"].unique()
```

```python
activated_jobs_cleaning.head()
```

```python
activated_jobs_cleaning.to_csv('Hiring_Computer_Science_Cleaning.csv', index = False, header = True)
```

```python
activated_jobs_transformation = activated_jobs_cleaning.copy()
activated_jobs_transformation = activated_jobs_cleaning[['DATA', 'SETTOREECONOMICO', 'PROVINCIAIMPRESA']]
```

## Data Samples Creation

```python
new_df = activated_jobs_transformation.copy()
new_df_2 = activated_jobs_transformation.copy()
Milano_Activated = activated_jobs_transformation.copy()
Bergamo_Activated = activated_jobs_transformation.copy()
```

```python
# Dataset containing total hirings of two sectors grouped by DATA and SETTOREECONOMICO
new_df = new_df.groupby(["DATA","SETTOREECONOMICO"]).count()
new_df["ATTIVATO"] = new_df["PROVINCIAIMPRESA"]
new_df = new_df.drop(["PROVINCIAIMPRESA"], axis = 1)
new_df
```

```python
# Creation of Dataset where there is only one SETTOREECONOMICO that includes others 
new_df_2 = new_df_2.replace(["Consulenza nel settore delle tecnologie dell'informatica","Altre attività dei servizi connessi alle tecnologie dell'informatica nca"], "Tecnologie Informatiche")
```

```python
# Dataset contained total hirings of Milano and Bergamo grouped by DATA and SETTOREECONOMICO
new_df_2 = new_df_2.groupby(["DATA", "SETTOREECONOMICO"]).count()
new_df_2["ATTIVATO"] = new_df_2["PROVINCIAIMPRESA"]
new_df_2 = new_df_2.drop(["PROVINCIAIMPRESA"], axis = 1)
new_df_2
```

```python
# Creation of two new datasets to compare results of models 
new_df.to_csv('Hiring_Computer_Science_Multiple_Sectors.csv', index = True, header = True)
new_df_2.to_csv('Hiring_Computer_Science_One_Sector.csv', index = True, header = True)
```
